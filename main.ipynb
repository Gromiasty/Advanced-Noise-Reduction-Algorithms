## Additive Noise

```python
# Load images from files
imagePeople = cv2.imread('./images/people.png', cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
imageDogs = cv2.imread('./images/dogs.png', cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
imageColors = cv2.imread('./images/colors.png', cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0

# Function to add Gaussian noise
def add_gaussian_noise(image, sigma):
    noise = np.random.normal(0, sigma, image.shape)
    noisy_image = image + noise
    noisy_image = np.clip(noisy_image, 0, 1)
    return noisy_image

# List of sigma values
sigma_values = [0.05, 0.1, 0.2, 0.3, 1]

# Plotting original and noisy images
fig, axes = plt.subplots(3, len(sigma_values) + 1, figsize=(15, 6))
axes[0, 0].imshow(imagePeople, cmap='gray')
axes[0, 0].set_title("Original People Image")
axes[0, 0].axis('off')

axes[1, 0].imshow(imageDogs, cmap='gray')
axes[1, 0].set_title("Original Dogs Image")
axes[1, 0].axis('off')

axes[2, 0].imshow(imageColors, cmap='gray')
axes[2, 0].set_title("Original Colors Image")
axes[2, 0].axis('off')

# Add noise and display images
for i, sigma in enumerate(sigma_values):
    noisy_image1 = add_gaussian_noise(imagePeople, sigma)
    noisy_image2 = add_gaussian_noise(imageDogs, sigma)
    noisy_image3 = add_gaussian_noise(imageColors, sigma)

    # Display noisy images
    axes[0, i + 1].imshow(noisy_image1, cmap='gray')
    axes[0, i + 1].set_title(f"Sigma {sigma}")
    axes[0, i + 1].axis('off')

    axes[1, i + 1].imshow(noisy_image2, cmap='gray')
    axes[1, i + 1].set_title(f"Sigma {sigma}")
    axes[1, i + 1].axis('off')

    axes[2, i + 1].imshow(noisy_image3, cmap='gray')
    axes[2, i + 1].set_title(f"Sigma {sigma}")
    axes[2, i + 1].axis('off')

plt.tight_layout()
plt.show()
```
